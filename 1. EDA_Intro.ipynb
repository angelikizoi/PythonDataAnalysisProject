```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  
import seaborn as sns

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df_gr_da = df[(df.job_country == 'Greece') & (df.job_title_short == 'Data Analyst')]
```

```python
df_plot = df_gr_da['job_location'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Locations in Greece for Data Analysts')
plt.ylabel('')
plt.xlabel('Counts of Job Locations')
plt.show()
```

```python
df_plot = df_gr_da['company_name'].value_counts().head(20).to_frame()

sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Companies in Greece for Data Analysts')
plt.ylabel('')
plt.xlabel('Counts of Companies')
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12, 5))

for i, (column, title) in enumerate(dict_column.items()):
    values = df_gr_da[column].value_counts()
    labels = values.index
    ax[i].pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

# plt.suptitle('Benefit Analysis of Data Jobs', fontsize=16)
plt.show()
```
